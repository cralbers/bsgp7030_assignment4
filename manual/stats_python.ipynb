## Hypothesis testing: comparing two groups
### Student's t-test
#### One sample tests: testing the value of a population mean
1-sample t-tests should be used on datasets that have a normal distribution, and the Wilcoxon test should be used on datasets that do not follow a normal distribution (non-parametric model). These tests are used to test the null hypothesis that the mean of a population is equal to some given value. 

````python
# test the null hypothesis that the mean of the population is equal to a given value
# assuming a normal distribution
ttest = sp.stats.ttest_1samp(data['VIQ'], 0)
print(f'T-test result: {ttest}')


# test the null hypothesis that the mean of the population is equal to a given value 
# assuming a not normal distribution 
wilcoxon = sp.stats.wilcoxon(data['VIQ'])
print(f'Wilcoxon test result: {wilcoxon}')
````

#### Two-sample t-test: testing for difference across populations
2-sample t-tests should be used on datasets that have a normal distribution, and the Mann-Whitney test should be used on datasets that do not follow a normal distribution (non-parametric model). These tests are used to test the null hypothesis that the mean of one population is equal to the mean of another population.

````python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']

# test whether the population means of two different populations are significantly different
# assuming a normal distribution
print(f'Two-sample t-test result: {sp.stats.ttest_ind(female_viq, male_viq)}')

# test whether the population means of two different populations are significantly different 
# with the assumption that the data does not follow a normal distribution
print(f'Mann-Whitney non-parametric test result: {sp.stats.mannwhitneyu(female_viq, male_viq)}')
````

### Paired tests: repeated measurements on the same individuals

````python
# Box plot of FSIQ and PIQ (different measures of IQ)
plt.figure(figsize=(4, 3))
data.boxplot(column=["FSIQ", "PIQ"])
plt.title('FSIQ and PIQ distributions')
````

````python
# test whether FSIQ and PIQ are significantly different using unpaired test (not correct)
unpaired = sp.stats.ttest_ind(data['FSIQ'], data['PIQ'])
print(f'Unpaired test result: {unpaired}')
````

The unpaired test is not correct for this dataset because it ignores the fact that FSIQ and PIQ are measured on the same individuals. To account for these repeated measures, a paired test must be used to remove inter-subject variability.

````python
# test whether FSIQ and PIQ are significantly different using paired test 
paired = sp.stats.ttest_rel(data['FSIQ'], data['PIQ'])
print(f'Paired test result: {paired}')
````

````python
# Boxplot of the difference between the FSIQ and PIQ measurements
plt.figure(figsize=(4, 3))
plt.boxplot(data["FSIQ"] - data["PIQ"])
plt.xticks((1,), ("FSIQ - PIQ",))
````

The paired test is equivalent to a one-sample t-test on the difference between paired observations. As such, either a one-sample t-test (for normally distributed data) or a non-parametric one-sample test (Wilcoxon) can be conducted on the difference between paired observations.

````python
# Paired test on 2 sets of data (ex init and final) is 
# equivalent to a one-sample test on the differences between paired observations
print(f'One sample t-test on difference: {sp.stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)}')

# non-parametric version
print(f'Wilcoxon test on difference: {sp.stats.wilcoxon(data['FSIQ'], data['PIQ'], method="approx")}')
````

## Linear models, multiple factors, and analysis of variance
### “formulas” to specify statistical models in Python
#### A simple linear regression
Test the hypothesis that y is a linear function of x:
$$y=x*coef+intercept+e $$
where e is observation noise

1. Fit a linear model using ordinary least squares
2. Test that the coefficient is not zero


````python
x = np.linspace(-5, 5, 20)

# To get reproducible values, provide a seed value
rng = np.random.default_rng(27446968)

# normal distributed noise
y = -5 + 3 * x + 4 * rng.normal(size=x.shape)

# Create a data frame containing all the relevant variables
data = pd.DataFrame({'x': x, 'y': y})

# Plot the data
plt.figure(figsize=(5, 4))
plt.plot(x, y, "o");
````

````python
# specify and fit an OLS model
model = smf.ols("y ~ x", data).fit()
print(model.summary())

sns.lmplot(x='x', y='y', data=data, ci=95)
plt.title('OLS Linear Regression')
plt.show()
````

## Testing for interactions

````python
sns.lmplot(y="wage", x="education", hue="sex", data=data)
````

````python
# do education and sex separately contribute to wage
result = smf.ols(formula="wage ~ education + sex", data=data).fit()
result.summary()
````

Now we look at whether wages increase more with education for males than females. To do this, add an interaction term (denoted by `*`) to the model to test for variance of slope across the two populations.

````python
# test whether wages increase more with education for males than females
result = smf.ols(formula='wage ~ education + sex + education * sex',
                 data=data).fit()
result.summary()
````
